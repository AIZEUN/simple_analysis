### 불량률이 2번째로 높은 고객 검사 데이터 확인

```python
df_costom = df1.loc[df1["불량구분"]=="고객검사"]
df_costom
```

```python
pd.pivot_table(df_costom,index=["불량구분","자재그룹"],values="불량수량",aggfunc="sum",margins=True)
```

## 불량그룹별 자재그룹의 불량수량 비교

```python
#한글 폰트사용
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
df_section = df1.loc[df1["불량구분"]=="공정검사"]
df_import = df1.loc[df1["불량구분"]=="수입검사"]
df_re = df1.loc[df1["불량구분"]=="자주검사"]
```

```python
for i in range(0,len(df_costom)):
    if pd.isna(df_costom.iloc[i,-2]):
        df_costom.iloc[i,-2] = "null"
for i in range(0,len(df_section)):
    if pd.isna(df_section.iloc[i,-2]):
        df_section.iloc[i,-2] = "null"
for i in range(0,len(df_import)):
    if pd.isna(df_import.iloc[i,-2]):
        df_import.iloc[i,-2] = "null"
for i in range(0,len(df_re)):
    if pd.isna(df_re.iloc[i,-2]):
        df_re.iloc[i,-2] = "null"
```

```python
plt.bar(df_costom["자재그룹"],df_costom["불량수량"],label="고객검사")
plt.bar(df_section["자재그룹"],df_section["불량수량"],label="공정검사")
plt.bar(df_import["자재그룹"],df_import["불량수량"],label="수입검사")
plt.bar(df_re["자재그룹"],df_re["불량수량"],label="자주검사")
plt.legend()
plt.show()
```
